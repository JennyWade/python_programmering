```python
#Splitting the vaccines left me with 7 columns but the vaccines were repeated within the columns, so I split the columns again
vaccine_list = covid_vaccinations["vaccines"].str.split(", ", expand=True)
vl0 = vaccine_list[0].str.split(",")
vl1 = vaccine_list[1].str.split(",")
vl2 = vaccine_list[2].str.split(",")
vl3 = vaccine_list[3].str.split(",")
vl4 = vaccine_list[4].str.split(",")
vl5 = vaccine_list[5].str.split(",")
vl6 = vaccine_list[6].str.split(",")
```

```python
#Below I've combined all series and drop duplicates to get a list with all unique vaccines
vl = pd.concat([vl0,vl1,vl2,vl3,vl4,vl5,vl6])
vl = vl.drop_duplicates()
vl
```
